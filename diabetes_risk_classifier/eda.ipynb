```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/diabetes.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# Outcome balance
sns.countplot(data=df, x='Outcome',hue='Outcome', palette='pastel', legend=False)
plt.title('Diabetes Outcome Distribution')
plt.xticks([0, 1], ['No Diabetes', 'Diabetic'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
important_features = ['Glucose', 'BMI', 'Age', 'Insulin']

for feature in important_features:
    sns.boxplot(x='Outcome', y=feature, hue='Outcome', data=df, palette='Set2', legend=False)
    plt.title(f'{feature} vs Outcome')
    plt.xticks([0, 1], ['No Diabetes', 'Diabetic'])
    plt.show()
```
