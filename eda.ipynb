```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Histogram: Avg Daily Usage Hours

sns.histplot(df['Avg_Daily_Usage_Hours'], kde=True, color='skyblue')
plt.title("Distribution of Daily Social Media Usage")
plt.xlabel("Hours per Day")
plt.ylabel("Number of Students")
plt.show()
```

```python
#  Boxplot: Addicted Score by Gender

sns.boxplot(x='Gender', y='Addicted_Score', data=df, hue='Gender', palette='Set2', legend=False)
plt.title("Addicted Score by Gender")
plt.show()
```

```python
#  Countplot: Most Used Platform

sns.countplot(x='Most_Used_Platform', data=df, hue='Most_Used_Platform', palette='pastel', legend=False)
plt.title("Most Popular Social Media Platforms")
plt.xticks(rotation=45)
plt.show()
```

```python
# Academic Impact vs Gender

sns.countplot(x='Affects_Academic_Performance', hue='Gender', data=df, palette='coolwarm')
plt.title("Addiction Impact on Academic Performance")
plt.show()
```

```python
# Correlation Heatmap

df_encoded = df.copy()
for col in df_encoded.select_dtypes(include='object'):
    # converts column str data to int...like yes,no to 1,0 
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

plt.figure(figsize=(12,6))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

```

```python
# Sleep vs Addiction (Scatter Plot)

sns.scatterplot(x='Sleep_Hours_Per_Night', y='Addicted_Score', hue='Gender', data=df)
plt.title("Sleep Hours vs Addicted Score")
plt.show()
```
